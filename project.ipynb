#### Exercise 01. Basic data analysis with native Python structures.  
Create a list with the grades of 5 students: [8.5, 9.2, 7.8, 8.9, 10].

- Calculate the average of the grades.

```python
# Create a list with grades
grades = [8.5, 9.2, 7.8, 8.9, 10]

#Calculate average
average = sum(grades)/len(grades)

print(f"{average:.2f}")
```

- Find the highest and lowest grade.

```python
# Find the Highest grade
highest = max(grades)

#Find the Lowest Grade
lowest = min(grades)

print(f"Highest Grade: {highest:.1f}")
print(f"Lowest Grade: {lowest:.1f}")
```

#### Exercise 01. What is the most expensive house in the entire dataset? (★☆☆)

Print the address and price of the selected house. For example:

`The house located at Calle del Prado, Nº20 is the most expensive, and its price is 5000000 USD.`

```python
#Find the most expensive house
most_expensive_house = ds.loc[ds['price'].idxmax()]

#Extract address and price from the most expensive house
address = most_expensive_house['address']
price = most_expensive_house['price']

print(f"The house located at {address} is the most expensive, and its price is {price} USD")
```

#### Exercise 02. What is the cheapest house in the dataset? (★☆☆)

This exercise is similar to the previous one, except now we are looking for the house with the lowest price. Remember to print the address and price of the selected house. For example:

`The house located at Calle Alcalá, Nº58 is the cheapest, and its price is 12000 USD.`


```python
#Find the cheapest house
cheapest_house = ds.loc[ds['price'].idxmin()]

#Extract address and price from the cheapest house
address = cheapest_house['address']
price = cheapest_house['price']

print(f"The house located at {address} is the cheapest, and its price is {price} USD")
```

#### Exercise 04. How many unique populations are in the dataset? (★☆☆)

Count the number of unique populations in the 'level5' column and print the names of the populations separated by commas. For example:

`> print(populations)`

`population1, population2, population3, ...`

#### Exercise 06. Remove the null values (NAs) from the dataset, if applicable (★★☆)

After removing the null values, compare the size of the DataFrame before and after the removal.

```python
#Check original size
original_size = ds.shape
print(f"Original dimension (Rows, Columns): {original_size}")

#Remove rows with NAs
clean_ds = ds.dropna()
clean_ds_rows = ds.dropna(axis=0)
clean_ds_columns = ds.dropna(axis=1)

print(f"-- CLEANING --")

#Clean DataFrame dimensions
after_deletions = clean_ds.shape

#Clean DataFrame dimensions after removing NAs from rows only
after_deletions_rows = clean_ds_rows.shape

#Clean DataFrame dimensions after removing NAs from columns only
after_deletions_columns = clean_ds_columns.shape

#Print deletions data
print("DataFrame after NAs deletions from both, rows and columns:", after_deletions)
print("DataFrame after NAs deletions from rows:", after_deletions_rows)
print("DataFrame after NAs deletions from columns:", after_deletions_columns)
```

#### Exercise 07. What is the average price in the population of "Arroyomolinos (Madrid)"? (★★☆)

Print the value obtained from the 'level5' column.

```python
# Get Arroyomolinos population info
arroyomolinos_info = ds[ds['level5'] == "Arroyomolinos (Madrid)"]

#Calculate and print Mean of price column for Arroyomolinos
mean_price = arroyomolinos_info['price'].mean()

print(f"The mean price in Arroyomolinos population {mean_price:.2f}")
```

#### Exercise 08. Plot the histogram of prices for the population of "Arroyomolinos (Madrid)" and explain what you observe (★★☆)

Print the histogram of the prices and write a brief analysis of the plot in the Markdown cell.

```python
import matplotlib.pyplot as plt

# Plot the histogram of prices
#Get only Arroyomolinos information population
arroyomolinos_info = ds[ds['level5'] == "Arroyomolinos (Madrid)"]

plt.hist(arroyomolinos_info['price'],bins=25, edgecolor= "white")
plt.xlabel('Price')
plt.ylabel('Number of properties')
plt.title('Histogram of Prices in Arroyomolinos (Madrid)')
plt.show
```
